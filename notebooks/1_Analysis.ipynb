```python
zero_coords = (data['pickup_longitude'] == 0) | (data['pickup_latitude'] == 0) | \
              (data['dropoff_longitude'] == 0) | (data['dropoff_latitude'] == 0)
print(f"Zero coordinates: {zero_coords.sum()} rows")

# 2. Extreme outliers (way outside NYC metro area)
extremes = (data['pickup_longitude'] < -75) | (data['pickup_longitude'] > -72) | \
           (data['pickup_latitude'] < 39) | (data['pickup_latitude'] > 42) | \
           (data['dropoff_longitude'] < -75) | (data['dropoff_longitude'] > -72) | \
           (data['dropoff_latitude'] < 39) | (data['dropoff_latitude'] > 42)
print(f"Extreme outliers: {extremes.sum()} rows")

# 3. Pickup == dropoff coordinates (with trip_duration > 0)
same_coords = (data['pickup_longitude'] == data['dropoff_longitude']) & \
              (data['pickup_latitude'] == data['dropoff_latitude'])
print(f"Pickup == dropoff: {same_coords.sum()} rows")
```

```python
same_loc = data[
    (data['pickup_latitude'] == data['dropoff_latitude']) &
    (data['pickup_longitude'] == data['dropoff_longitude'])
]

same_loc['trip_duration'].describe()
```

```python
import pandas as pd
import numpy as np

# Combine pickup date and time
data["pickup_datetime"] = pd.to_datetime(
    data["pickup_date"].astype(str) + " " + data["pickup_time"].astype(str)
)

# Extract useful features
data["pickup_month"] = data["pickup_datetime"].dt.month
data["pickup_weekday"] = data["pickup_datetime"].dt.weekday
data["pickup_hour"] = data["pickup_datetime"].dt.hour

# Cyclic encoding for day of month
data["pickup_day_sin"] = np.sin(
    2 * np.pi * data["pickup_datetime"].dt.day / 31
)

data["pickup_day_cos"] = np.cos(
    2 * np.pi * data["pickup_datetime"].dt.day / 31
)

# Cyclic encoding for hour of day
data["pickup_hour_sin"] = np.sin(
    2 * np.pi * data["pickup_hour"] / 24
)

data["pickup_hour_cos"] = np.cos(
    2 * np.pi * data["pickup_hour"] / 24
)

# Remove original columns
data.drop(
    columns=[
        "pickup_datetime",
        "pickup_date",
        "pickup_time",
        "pickup_hour",
    ],
    inplace=True,
)

# Preview
data.sample(5)
```
