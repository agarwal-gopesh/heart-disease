```python
data['pickup_datetime'] = pd.to_datetime(data['pickup_datetime'])
data['pickup_date'] = data['pickup_datetime'].dt.date
data['pickup_time'] = data['pickup_datetime'].dt.time
cols = ['vendor_id', 'pickup_date', 'pickup_time'] + [c for c in data.columns if c not in ['vendor_id', 'pickup_date', 'pickup_time', 'pickup_datetime']]
data = data[cols]
data.sample(10)
```
